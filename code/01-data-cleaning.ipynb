### Import libraries and data

```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import warnings
warnings.filterwarnings('ignore')
from pandas_profiling import ProfileReport
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

%matplotlib inline
```

```python
#read csv
solar_ctry = pd.read_csv('../data/EMHIRESPV_TSh_CF_Country_19862015.csv')
solar_nuts = pd.read_csv('../data/EMHIRES_PVGIS_TSh_CF_n2_19862015.csv')

#preview data
print("---------------------------- Top 5 rows of solar_ctry -------------------------")
display(solar_ctry.head())
print(f'Shape of solar_ctry : {solar_ctry.shape}')
print("----------------------------- Data types in solar_ctry ------------------------")
print(solar_ctry.dtypes)
print("---------------------------- Top 5 rows of solar_nuts -------------------------")
display(solar_nuts.head())
print(f'Shape of solar_nuts : {solar_nuts.shape}')
print("----------------------------- Data types in solar_nuts ------------------------")
print(solar_nuts.dtypes)
```

**Observations :**

Data columns in solar energy potential data by country refer to the various countries in the EU region whereas for columns in solar energy potential data by NUTS system, refer to the different regions categorised using the NUTS system. 

The solar energy potential data by NUTS system has a column `time_step` which is not relevant since the indices can be used to assign datetime. So I will drop this. 

For both dataframnes, the values reflect the hourly estimates of the area's solar energy potential from 1986 to 2015. This explains why the first 5 rows had nil values as they reflect the wee hours of the morning prior to sunset. 

All values are already in floats or integers, so there is no need to change the data types. 

From the above, I'd perform the following :

- On `solar_ctry` and `solar_nuts`, **add a time column** to **track which hours of the day** would the area's energy potential be higher than others. I expect there to be a spike in the afternoon hours. There may also be seasonality across the years, so I will also track the month and week.  


- drop `time_step` column in `solar_nuts`


- Change `CY` data type in `solar_ctry` from integer to float


- For `solar_ctry`, **replace country codes with country names for readability**


```python
def add_time(df):
    '''adds time column based on start and end year'''
    df['time'] = pd.date_range(start='1/1/1986', periods=df.shape[0],freq='H')
    df['hour'] = df['time'].dt.hour
    df['week'] = df['time'].dt.week
    df['month'] = df['time'].dt.month
    #check
    print("-------------- Top 5 rows ----------------")
    display(df.head())
    print("------------------ Last 5 rows -----------")
    display(df.tail())
```

```python
def change_ctry_name(df):
    '''function that changes country code to its name'''
    country_name = {'AT': 'Austria',
                    'BE': 'Belgium',
                    'BG': 'Bulgaria',
                    'CH': 'Switzerland',
                    'CY': 'Cyprus',
                    'CZ': 'Czech Republic',
                    'DE': 'Germany',
                    'DK': 'Denmark',
                    'EE': 'Estonia',
                    'ES': 'Spain',
                    'FI': 'Finland',
                    'FR': 'France',
                    'EL': 'Greece',
                    'HR': 'Croatia',
                    'HU': 'Hungary',
                    'IE': 'Ireland',
                    'IT': 'Italy',
                    'LT': 'Lithuania',
                    'LU': 'Luxembourg',
                    'LV': 'Latvia',
                    'NL': 'Netherlands',
                    'NO': 'Norway',
                    'PL': 'Poland',
                    'PT': 'Portugal',
                    'RO': 'Romania',
                    'SI': 'Slovenia',
                    'SK': 'Slovakia',
                    'SE': 'Sweden',
                    'UK': 'United Kingdom'}
    df.columns = [country_name[j] for j in df.columns]
```

```python
#apply function that changes column names
change_ctry_name(solar_ctry)
```

```python
#apply function that adds time columns and its variants
add_time(solar_ctry)
add_time(solar_nuts)
```

```python
#drop time_step column in solar_nuts
solar_nuts.drop(columns = 'time_step', inplace=True)
```

```python
#change type from integer to float
solar_ctry['Cyprus'] = solar_ctry['Cyprus'].astype(float)
```

### Preliminary review of data
Generate a pandas profiling report for a preliminary review of the data set

```python
#generate profile report about solar energy potential by country
solar_ctry_report = ProfileReport(solar_ctry)
solar_ctry_report.to_widgets()
```

From the report for `solar_ctry`, I note the following: 

- there are no missing values
- data for each country is heavily right-skewed (ie. there are a lot of instances where no energy is generated) 
- there is high correlation between the countries, which is expected since they are pretty close together so the exposure to sun is relatively similar. Analysis can be done in clusters where countries exposed to similar levels of radiance be grouped together and further analysis done on one country for each cluster.
- there are no negative values so data collected is pretty clean

```python
#generate profile report about solar energy potential by NUTS
solar_nuts_report = ProfileReport(solar_nuts, minimal=True) #minimal argument creates a simplified report. 
solar_nuts_report.to_widgets()
```

From the report for `solar_nuts`, I note the following:

- Letters represent the respective countries, and the numbers represent the regions based on the [NUTS 2 system](https://en.wikipedia.org/wiki/Nomenclature_of_Territorial_Units_for_Statistics)
- Values in this dataset is similar to the ones by country where they are heavily right-skewed. 
- there are also no negative values in this dataset.
